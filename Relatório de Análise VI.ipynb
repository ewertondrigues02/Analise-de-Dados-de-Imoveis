**Criando Novos Campos**

```python
# Importar o pandas com apelido pd
import pandas as pd

# Importa o CSV com os dados de imóveis residenciais
imoveis = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Analise de Dados/Análise de Dados de imóveis/imoveis-resienciais.csv', sep=';')

# Exibe as cincos primeiras linhas do dataframe
imoveis.head()
```

```python
imoveis.shape[0]
```

```python
# Cria uma coluna para o valor bruto mensal a ser pago pelo imóvel
imoveis['Valor Bruto'] = imoveis['Valor'] + imoveis['Condominio'] + imoveis['IPTU']
imoveis.head(10)
```

```python
# Cria uma coluna com o valor total do aluguel por metro quadrado do imovel
imoveis['Valor m²'] = (imoveis['Valor'] / imoveis['Area']).round(2)
imoveis.head(10)
```

```python
# Cria uma coluna para indicar o valor bruto do metro quadrado
imoveis['Valor Bruto m²'] = (imoveis['Valor Bruto'] / imoveis['Area']).round(2)
imoveis.head(10)
```

```python
# Obtém uma série com os tipos únicos da coluna tipo
imoveis['Tipo'].drop_duplicates()
```

```python
tipos_de_casas = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
tipos_de_casas
```

```python
# Aplica uma função mapeadora na coluna de tipo do imóvel
# Para cada valor da coluna 'Tipo', a função retorna ou 'Casa' ou 'Apartamento'
# Se o imóvel for um dos tipos de casa, retorna 'Casa'
# Caso contrário, retorna 'Apartamento'
imoveis['Tipo'].map(lambda x: 'Casa' if x in tipos_de_casas else 'Apartamento')
```

```python
# Cria uma nova coluna no dataframe
# Indica qual o tipo do imóvel: ou Casa ou Apartamento
imoveis['Tipo Agregado'] = imoveis['Tipo'].map(lambda x: 'Casa' if x in tipos_de_casas else 'Apartamento')
imoveis.head(10)
```

```python
# Remove as colunas criadas
imoveis.drop(['Valor m²', 'Valor Bruto m²'], axis=1, inplace=True)
imoveis.head(10)
```

```python
# Exporta a base de dados
imoveis.to_csv('imoveis-residenciais.csv', sep=';', index=False)
```
